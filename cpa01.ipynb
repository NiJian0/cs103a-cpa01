# Analysis of historical stock prices (last 5 years) for all companies currently found on the S&P 500 index.
The data comes from kaggle and the site for this particular data is [https://www.kaggle.com/datasets/camnugent/sandp500].
I've downloaded the data and stored it in all_stocks_5yr.csv

Some questions I try to answer

how is the stock closing price of AAPL changed over years
what information we can get from the closing price of stock 'AAPL', 'MSFT' and 'GOOGL' changing over years
what is the candlestick chart of AAPL

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import plotly.graph_objects as go
import mplfinance as fplt
```

```python
df = pd.read_csv('all_stocks_5yr.csv')
df
```

```python
df.describe()
```

```python
df.columns
```

```python
df[:1260].plot(figsize=(15,10))
plt.grid()
plt.title('the stock information of AAL',fontsize=16)
```

# Create a pivot table

index is the date, and columns are the values in the Name column the value is the closing price. We can then plot just the columns we want!



```python
df1 = pd.pivot_table(df, values='close', index=["date"], columns=["Name"])
df1
```

```python
df1['AAPL'].plot(figsize=(15,10))
plt.grid()
plt.title('the closing price of AAPL',fontsize=16)
```

# Calculate the total volumes for all stocks
We can use the groupby method to determine sum up the volumes of all stocks so far and show the top 20

```python
df2 = df.groupby('Name')['volume'].agg(np.sum).sort_values(ascending=False).head(20)
df2
```

```python
df1[['AAPL', 'MSFT', 'GOOGL']].plot(subplots=True)
```

```python
df.index = pd.DatetimeIndex(df['date'])
df3 = df[df['Name']=='AAPL']
fplt.plot(
            df3,
            type='candle',
            style='charles',
            title='Apple, March - 2020',
            ylabel='Price ($)',
            )
```
